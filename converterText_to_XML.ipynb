```python
def convert_row(row):
    return f"""<person>
    <id>{row.person_id}</id>
    <persName>
        <surname>{row.surname}</surname>
        <name>{row[2]}</name>
        <patronimic>{row.patronimic}</patronimic>
        <all_name_variations>{row.all_name_variations}</all_name_variations>
        <name_variations>{row.name_variations}</name_variations>
        <surname_variations>{row.surname_variations}</surname_variations>
        <patronimic_variations>{row.patronimic_variations}</patronimic_variations>
        <fullname_variations>{row.fullname_variations}</fullname_variations>
    </persName>
    <event type='birth'>
        <birth_year>{row.birth_year}</birth_year>
        <birth_place>{row.birth_place}</birth_place>
        <birth_year_comment>{row.birth_year_comment}</birth_year_comment>
    </event>
    <age>{row.age}</age>
    <sex>{row.sex}</sex>
    <nation>{row.nation}</nation>
    <citizenship>{row.citizenship}</citizenship>
    <occupation>{row.occupation}</occupation>
    <live_place>{row.live_place}</live_place>
    <education>{row.education}</education>
    <party>{row.party}</party>
    <family>{row.family}</family>
    <event type='arrest'>
        <arrest_date>{row.arest_date}</arrest_date>
        <arrest_organ>{row.arest_organ}</arrest_organ>
        <arrest_type>{row.arest_type}</arrest_type>
        <criminal_case>{row.criminal_case}</criminal_case>
        <judicial_organ>{row.judicial_organ}</judicial_organ>
        <process_date>{row.process_date}</process_date>
        <criminal_article>{row.criminal_article}</criminal_article>
        <sentence>{row.sentence}</sentence>
    </event>
    <event type='execution'>
        <was_executed>{row.was_executed}</was_executed>
        <death_place>{row.death_place}</death_place>
        <death_date>{row.death_date}</death_date>
    </event>
    <previous_repression>{row.previous_repression}</previous_repression>
    <next_repression>{row.next_repression}</next_repression>
    <event type='rehabilitation'>
        <rehabilitation_organ>{row.rehabilitation_organ}</rehabilitation_organ>
        <rehabilitation_reason>{row.rehabilitation_reason}</rehabilitation_reason>
        <rehabilitation_date>{row.rehabilitation_date}</rehabilitation_date>
    </event>
    <memory_book>{row.memory_book}</memory_book>
    <memory_book_url>{row.memory_book_url}</memory_book_url>
</person>&"""
```

```python
import lxml.etree as ET
from xml.dom.minidom import parseString
```

```python
def creating_xml_from_base_memo(table):
    plain_xml = '\n'.join(table.apply(convert_row, axis=1))
    for xml in plain_xml.split('&')[:-1]:
        tree = ET.XML(xml)
        person_id = tree.findtext('id')
        for tag in tree.iter():
            if tag.text == 'None':
                tag.getparent().remove(tag)
        xml_tree = ET.tostring(tree, encoding="unicode")
        xml_tree = '\n'.join([line for line in 
                              parseString(xml_tree).toprettyxml().split('\n') if line.strip()])
        print(xml_tree)
        with open(f'{person_id}.xml', 'w', encoding='utf-8') as f:
            print(xml_tree, file=f)
```
